```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Conv2D, BatchNormalization, MaxPooling2D, Activation, Flatten, Dropout, Concatenate
from tensorflow.keras.optimizers import SGD, Adam
#from keras.preprocessing.image import img_to_array
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import ResNet50V2
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam, SGD
from keras.layers import Dense, Conv2D, BatchNormalization, MaxPooling2D, Activation, Flatten, Dropout, Input
import matplotlib
matplotlib.use("Agg")
import scipy.io
import glob
# import the necessary packages
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import img_to_array
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
#from pyimagesearch.smallervggnet import SmallerVGGNet
from keras.models import Sequential
from keras.layers import Dense, Conv2D, BatchNormalization, MaxPooling2D, Activation, Flatten, Dropout
import matplotlib.pyplot as plt
from imutils import paths
import numpy as np
import argparse
import random
import pickle
import cv2
import os
from sklearn.model_selection import cross_val_score
from tensorflow.keras.layers import concatenate
def complete_base():
  #inputs
  Input_Color = Input(shape=(96, 96, 3))
  Input_SFDI = Input(shape=(211,211,8))

  #Color Image Branch
  color = Conv2D(16, (3, 3), padding="same")(Input_Color)
  color = Activation("relu")(color)
  color = BatchNormalization(axis=-1)(color)
  color = Conv2D(16, (3, 3), padding="same")(color)
  color = Activation("relu")(color)
  color = BatchNormalization(axis=-1)(color)
  color = MaxPooling2D(pool_size=(2, 2), strides=(2,2))(color)
  color = Dropout(0.25)(color)

  color = Conv2D(32, (3, 3), padding="same")(color)
  color = Activation("relu")(color)
  color = BatchNormalization(axis=-1)(color)
  color = Conv2D(32, (3, 3), padding="same")(color)
  color = Activation("relu")(color)
  color = BatchNormalization(axis=-1)(color)
  color = MaxPooling2D(pool_size=(2, 2), strides=(2,2))(color)
  color = Dropout(0.25)(color)
  
  color = Flatten()(color)
  color = Model(inputs=Input_Color, outputs=color)

  sfdi = Conv2D(16, (3, 3), padding="same")(Input_SFDI)
  sfdi = Activation("relu")(sfdi)
  sfdi = BatchNormalization(axis=-1)(sfdi)
  sfdi = Conv2D(16, (3, 3), padding="same")(sfdi)
  sfdi = Activation("relu")(sfdi)
  sfdi = BatchNormalization(axis=-1)(sfdi)
  sfdi = MaxPooling2D(pool_size=(2, 2), strides=(2,2))(sfdi)
  sfdi = Dropout(0.25)(sfdi)

  sfdi = Conv2D(32, (3, 3), padding="same")(sfdi)
  sfdi = Activation("relu")(sfdi)
  sfdi = BatchNormalization(axis=-1)(sfdi)
  sfdi = Conv2D(32, (3, 3), padding="same")(sfdi)
  sfdi = Activation("relu")(sfdi)
  sfdi = BatchNormalization(axis=-1)(sfdi)
  sfdi = MaxPooling2D(pool_size=(2, 2), strides=(2,2))(sfdi)
  sfdi = Dropout(0.25)(sfdi)
  
  sfdi = Flatten()(sfdi)
  sfdi = Model(inputs=Input_SFDI, outputs=sfdi)
  
  combined = concatenate([color.output, sfdi.output])

  combined = Dense(128)(combined)
  combined = Activation("relu")(combined)
  combined = BatchNormalization()(combined)
  combined = Dropout(0.5)(combined)
  combined = Dense(1)(combined)
  combined = Activation("sigmoid")(combined)

  model = Model(inputs = [color.input, sfdi.input], outputs = combined)
  opt = Adam(lr=0.01)
  model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])
  print(model.summary())
  return model
model = complete_base()
(train_colorX, test_colorX, train_colorY, test_colorY) = train_test_split(color_data, color_labels, test_size=0.2, random_state=42)
(train_sfdiX, test_sfdiX, train_sfdiY, test_sfdiY) = train_test_split(sfdi_data, sfdi_labels, test_size=0.2, random_state=42)
print(train_colorY[48])
print(train_colorY[148])
print(train_colorY[248])
print(train_colorY[348])
print(train_colorY[448])

print(train_sfdiY[48])
print(train_sfdiY[148])
print(train_sfdiY[248])
print(train_sfdiY[348])
print(train_sfdiY[448])

H = model.fit(x = [train_colorX, train_sfdiX], y = train_colorY, batch_size = 80, validation_data = ([test_colorX, test_sfdiX], test_colorY), steps_per_epoch=(len(train_colorX)//80), epochs=100, verbose=1)
```

Transfer Learning Model

```python
from keras.applications import ResNet50V2
def multify_weights(kernel, out_channels):
  mean_1d = np.mean(kernel, axis=-2).reshape(kernel[:,:,-1:,:].shape)
  tiled = np.tile(mean_1d, (out_channels, 1))
  return(tiled)


# Loop through layers of both original model 
# and custom model and copy over weights 
# layer_modify refers to first convolutional layer
def copy_weights_tl(model_orig, custom_model, layer_modify):
  layer_to_modify = [layer_modify]

  conf = custom_model.get_config()
  layer_names = [conf['layers'][x]['name'] for x in range(len(conf['layers']))]

  for layer in model_orig.layers:
    if layer.name in layer_names:
      if layer.get_weights() != []:
        target_layer = custom_model.get_layer(layer.name)

        if layer.name in layer_to_modify:    
          kernels = layer.get_weights()[0]
          biases  = layer.get_weights()[1]

          kernels_extra_channel = np.concatenate((kernels,
                                                  multify_weights(kernels, 8 - 3)),
                                                  axis=-2)
                                                  
          target_layer.set_weights([kernels_extra_channel, biases])
          target_layer.trainable = False

        else:
          target_layer.set_weights(layer.get_weights())
          target_layer.trainable = False
def combined_transfer():
  #######INPUTS#######
  inputs_color = Input((96,96,3))
  inputs_sfdi = Input((211,211,8))
  #######COLOR#######
  resnet50 = ResNet50V2(weights='imagenet', include_top=False)
  #resnet50.get_layer(name = 'resnet50v2').name='resnet50v2_1'

  config = resnet50.get_config()
  config["layers"][0]["config"]["batch_input_shape"] = (None, 96, 96, 3)
  config["layers"][2]["config"]["strides"] = (1, 1)

  x1 = resnet50(inputs_color)
  x1 = tf.keras.activations.relu(x1)
  x1 = Flatten()(x1)
  x1 = Model(inputs=inputs_color, outputs=x1)
  x1.get_layer(name = 'resnet50v2')._name='resnet50v2_1'
  #######SFDI#######
  resnet501 = ResNet50V2(weights='imagenet', include_top=False)

  config = resnet501.get_config()
  config["layers"][0]["config"]["batch_input_shape"] = (None, 211, 211, 8)
  config["layers"][2]["config"]["strides"] = (1, 1)

  resnet_custom = tf.keras.models.Model.from_config(config)
  modify_name = config["layers"][2]["config"]["name"]

  copy_weights_tl(resnet501, resnet_custom, modify_name)
  x2 = resnet_custom(inputs_sfdi)
  x2 = tf.keras.activations.relu(x2)
  x2 = Flatten()(x2)
  x2 = Model(inputs=inputs_sfdi, outputs=x2)

  conv4 = concatenate([x1.output, x2.output])

  conv4 = Dense(128)(conv4)
  conv4 = Activation("relu")(conv4)
  conv4 = BatchNormalization()(conv4)
  conv4 = Dropout(0.5)(conv4)
  conv4 = Dense(1)(conv4)
  conv4 = Activation("sigmoid")(conv4)
  model = Model(inputs = [x1.input, x2.input], outputs = conv4)
  opt = Adam(lr=0.01)
  model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])
  print(model.summary())
  return model
transfer_model = combined_transfer()
(train_colorX, test_colorX, train_colorY, test_colorY) = train_test_split(color_data, color_labels, test_size=0.2, random_state=42)
(train_sfdiX, test_sfdiX, train_sfdiY, test_sfdiY) = train_test_split(sfdi_data, sfdi_labels, test_size=0.2, random_state=42)
print(train_colorY[48])
print(train_colorY[148])
print(train_colorY[248])
print(train_colorY[348])
print(train_colorY[448])

print(train_sfdiY[48])
print(train_sfdiY[148])
print(train_sfdiY[248])
print(train_sfdiY[348])
print(train_sfdiY[448])

H = transfer_model.fit(x = [train_colorX, train_sfdiX], y = train_colorY, batch_size = 80, validation_data = ([test_colorX, test_sfdiX], test_colorY), steps_per_epoch=(len(train_colorX)//80), epochs=100, verbose=1)
```

```python
results = transfer_model.evaluate([color_test_data, sfdi_test_data], color_test_labels)
print(results)
```
